```python
spark.sql("CREATE SCHEMA IF NOT EXISTS telco")
```

```python
import os

# List all files under the Files/telco_data directory in the default Lakehouse
files = [f.name for f in notebookutils.fs.ls("Files/telco_data") if f.name.endswith(".csv")]

for file_name in files:
    table_name = os.path.splitext(file_name)[0]
    file_path = f"Files/telco_data/{file_name}"
    df = spark.read.option("header", True).option("inferSchema", True).csv(file_path)
    df.write.mode("overwrite").saveAsTable(f"telco.{table_name}")

# Optionally display the created tables
for file_name in files:
    table_name = os.path.splitext(file_name)[0]
    display(spark.table(f"telco.{table_name}"))
```
